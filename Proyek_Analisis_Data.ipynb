## Visualization & Explanatory Analysis

### Pertanyaan 1: Dari tahun 2011-2012, pada musim apa sepeda banyak disewa (baik oleh pengguna biasa/terdaftar)?

```python
season_sort = day_df.groupby("season")[["registered", "casual"]].sum().reset_index()

season_name = ["spring","summer", "fall", "winter"]

plt.figure(figsize=(10, 6))

plt.bar(
    season_sort["season"],
    season_sort["registered"],
    label="Registered",
    color="tab:blue"
)

plt.bar(
    season_sort["season"],
    season_sort["casual"],
    label="Casual",
    color="tab:green",
)

plt.xlabel(None)
plt.ylabel(None)
plt.title("Jumlah Penyewa Sepeda Berdasarkan Musim")
plt.legend()
plt.show()
```

### Pertanyaan 2:Dari tahun 2011-2012, pada cuaca apa sepeda banyak disewa (baik oleh pengguna biasa/terdaftar)?

```python
weathersit_sort = day_df.groupby("weathersit")[["registered", "casual"]].sum().reset_index()

weathersit_name = ["Clear/Few clouds/Partly cloudy", "Mist + Cloudy\ Mist", "Light Snow/Light Rain","Heavy Rain/ Snow"]

plt.figure(figsize=(10, 6))

plt.bar(
    weathersit_sort["weathersit"],
    weathersit_sort["registered"],
    label="Registered",
    color="tab:orange"
)

plt.bar(
    weathersit_sort["weathersit"],
    weathersit_sort["casual"],
    label="Casual",
    color="tab:green",
)

plt.xlabel(None)
plt.ylabel(None)
plt.title("Jumlah Penyewa Sepeda Berdasarkan Cauca")
plt.legend()
plt.show()
```

**Insight:**
- Diperoleh bahwa pada musim gugur (*fall*) memiliki jumlah penyewa sepeda terbanyak
- Diperoleh bahwa pada cuaca *Clear, Few clouds, Partly cloudy, Partly cloudy* memiliki jumlah penyewa sepeda terbanyak

## Conclusion

- Dari tahun 2011-2012, jumlah penyewa sepeda terbanyak berdasarkan musim adalah pada musim gugur (*fall*) dengan total jumlah pengguna biasa dan terdaftar sebanyak 1061129
- Dari tahun 2011-2012, jumlah penyewa sepeda terbanyak berdasarkan cuaca adalah pada cuaca *Clear, Few clouds, Partly cloudy, Partly cloudy* dengan total jumlah pengguna biasa dan terdaftar sebanyak 2257952
